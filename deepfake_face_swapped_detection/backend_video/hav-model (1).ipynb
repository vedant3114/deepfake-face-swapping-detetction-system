```python
# ===================================================================
# SECTION 1: IMPORTS AND SETUP
# ===================================================================

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, random_split
import torchaudio
import torchvision
from torchvision import transforms
import cv2
import pandas as pd
import numpy as np
import os
from pathlib import Path
import math
import random
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, confusion_matrix, classification_report
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
from collections import defaultdict
import warnings
from tqdm import tqdm
import time
warnings.filterwarnings('ignore')

print("=" * 80)
print("OPTIMIZED MULTIMODAL DEEPFAKE DETECTION SYSTEM")
print("ENHANCED FOR SMALL DATASETS WITH STRONG REGULARIZATION")
print("=" * 80)
print("KEY OPTIMIZATIONS:")
print("✓ Reduced model complexity (ResNet18, smaller hidden dims)")
print("✓ Dynamic class balancing for loss function")
print("✓ Strong regularization (high dropout, weight decay)")
print("✓ Label smoothing for generalization")
print("✓ Optimized hyperparameters for small dataset")
print("=" * 80)
print("Section 1: Imports and Setup - COMPLETED ✓")
print(f"PyTorch Version: {torch.__version__}")
print(f"TorchAudio Version: {torchaudio.__version__}")
print(f"TorchVision Version: {torchvision.__version__}")

# Device setup
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Device: {device}")
if torch.cuda.is_available():
    print(f"GPU: {torch.cuda.get_device_name(0)}")
    print(f"GPU Memory: {torch.cuda.get_device_properties(0).total_memory / 1024**3:.1f} GB")

# Hyperparameters - OPTIMIZED FOR SMALL DATASET
BATCH_SIZE = 8  # Increased for better gradient estimates
LEARNING_RATE = 5e-5  # Reduced for stability
NUM_EPOCHS = 50  # More epochs with early stopping
AUDIO_SAMPLE_RATE = 16000
VIDEO_FPS = 25
SEQUENCE_LENGTH = 32  # Use full cached sequence length
AUDIO_SEGMENT_LENGTH = 1.0  # seconds per segment
IMG_SIZE = 112  # REDUCED - Smaller input size
AUDIO_N_MELS = 64  # REDUCED - Less frequency resolution
AUDIO_N_FFT = 1024  # REDUCED
VALIDATION_SPLIT = 0.2
USE_PREPROCESSING = True  # Set to False to extract features on-the-fly (slower but uses less disk space)
CACHE_PATH_ROOT = "/kaggle/input/hav-df-cache"  # Root directory for cache folders. Change this to your existing cache location.

# STRONG REGULARIZATION PARAMETERS
DROPOUT_RATE = 0.5  # Increased dropout
WEIGHT_DECAY = 1e-2  # Stronger weight decay
LABEL_SMOOTHING = 0.1  # Label smoothing
HIDDEN_DIM = 128  # REDUCED - Much smaller hidden dimensions

print(f"Hyperparameters configured:")
print(f"  Batch Size: {BATCH_SIZE}")
print(f"  Learning Rate: {LEARNING_RATE}")
print(f"  Epochs: {NUM_EPOCHS}")
print(f"  Sequence Length: {SEQUENCE_LENGTH}")
print(f"  Image Size: {IMG_SIZE}")
print(f"  Cache Path Root: {CACHE_PATH_ROOT}")
print("=" * 80)

# ===================================================================
# SECTION 2: DATA LOADING AND PREPROCESSING
# ===================================================================

class MultiModalDeepfakeDataset(Dataset):
    def __init__(self, video_paths, labels, audio_transform=None, video_transform=None, mode='train', preprocess=True):
        """
        Novel Multi-Modal Dataset with temporal consistency checking
        """
        self.video_paths = video_paths
        self.labels = labels
        self.audio_transform = audio_transform
        self.video_transform = video_transform
        self.mode = mode
        self.preprocess = preprocess
        
        # Setup transforms
        self.setup_transforms()
        
        # Preprocessing cache
        self.cache_dir = os.path.join(CACHE_PATH_ROOT, f"cache_{self.mode}")
        self.audio_cache = {}
        self.video_cache = {}
        
        print(f"Dataset initialized in {self.mode} mode:")
        print(f"  Total videos: {len(video_paths)}")
        print(f"  Real videos: {sum(1 for l in labels if l == 0)}")
        print(f"  Fake videos: {sum(1 for l in labels if l == 1)}")
        
        # Preprocess data if enabled
        if self.preprocess and len(video_paths) > 0:
            self.preprocess_data()
        
    def setup_transforms(self):
        """Setup audio and video transforms"""
        # Audio transforms
        self.mel_spectrogram = torchaudio.transforms.MelSpectrogram(
            sample_rate=AUDIO_SAMPLE_RATE,
            n_fft=AUDIO_N_FFT,
            n_mels=AUDIO_N_MELS,
            hop_length=512,
            power=2.0
        )
        
        # Video transforms
        if self.mode == 'train':
            self.video_transform = transforms.Compose([
                transforms.ToPILImage(),
                transforms.Resize((IMG_SIZE, IMG_SIZE)),
                transforms.RandomHorizontalFlip(p=0.3),
                transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1),
                transforms.ToTensor(),
                transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
            ])
        else:
            self.video_transform = transforms.Compose([
                transforms.ToPILImage(),
                transforms.Resize((IMG_SIZE, IMG_SIZE)),
                transforms.ToTensor(),
                transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
            ])
    
    def preprocess_data(self):
        """Preprocess all videos to cache features for faster training"""
        print(f"\nPreprocessing {len(self.video_paths)} videos for {self.mode} set...")
        print("This may take a while but will speed up training significantly.")
        
        # Create cache directory
        os.makedirs(self.cache_dir, exist_ok=True)
        
        # Progress bar for preprocessing
        pbar = tqdm(enumerate(self.video_paths), total=len(self.video_paths), 
                   desc=f"Preprocessing {self.mode} videos")
        
        for idx, video_path in pbar:
            cache_file = os.path.join(self.cache_dir, f"video_{idx}.pt")
            
            if os.path.exists(cache_file):
                # Load from cache
                cached_data = torch.load(cache_file)
                self.audio_cache[idx] = cached_data['audio']
                self.video_cache[idx] = cached_data['video']
                pbar.set_postfix({"status": "loaded from cache"})
            else:
                # Extract features
                pbar.set_postfix({"status": "extracting features"})
                
                audio_features = self._extract_audio_features(video_path)
                video_features = self._extract_video_features(video_path)
                
                # Cache features
                self.audio_cache[idx] = audio_features
                self.video_cache[idx] = video_features
                
                # Save to disk cache
                torch.save({
                    'audio': audio_features,
                    'video': video_features
                }, cache_file)
                
                pbar.set_postfix({"status": "cached"})
        
        pbar.close()
        print(f"Preprocessing complete! Features cached in '{self.cache_dir}' directory.")
    
    def extract_audio_features(self, video_path):
        """Extract audio features using novel temporal segmentation"""
        # If preprocessing is enabled, this won't be called during training
        return self._extract_audio_features(video_path)
    
    def _extract_audio_features(self, video_path):
        """Internal method for audio feature extraction"""
        try:
            # Load audio from video
            cap = cv2.VideoCapture(str(video_path))
            total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
            fps = cap.get(cv2.CAP_PROP_FPS)
            cap.release()
            
            # Use torchaudio to extract audio from video
            waveform, sample_rate = torchaudio.load(str(video_path))
            
            # Resample if necessary
            if sample_rate != AUDIO_SAMPLE_RATE:
                resampler = torchaudio.transforms.Resample(sample_rate, AUDIO_SAMPLE_RATE)
                waveform = resampler(waveform)
            
            # Convert to mono if stereo
            if waveform.shape[0] > 1:
                waveform = torch.mean(waveform, dim=0, keepdim=True)
            
            # Segment audio into fixed-length chunks
            segment_samples = int(AUDIO_SAMPLE_RATE * AUDIO_SEGMENT_LENGTH)
            audio_segments = []
            
            for i in range(SEQUENCE_LENGTH):
                start_idx = i * segment_samples
                end_idx = start_idx + segment_samples
                
                if end_idx <= waveform.shape[1]:
                    segment = waveform[:, start_idx:end_idx]
                else:
                    # Pad or repeat if necessary
                    if start_idx < waveform.shape[1]:
                        segment = waveform[:, start_idx:]
                        padding_needed = segment_samples - segment.shape[1]
                        if padding_needed > 0:
                            segment = F.pad(segment, (0, padding_needed))
                    else:
                        # If we've run out of audio, create a zero segment
                        segment = torch.zeros(1, segment_samples)
                
                # Convert to mel spectrogram
                mel_spec = self.mel_spectrogram(segment)
                audio_segments.append(mel_spec.squeeze(0))
            
            # Ensure we have exactly SEQUENCE_LENGTH segments
            if len(audio_segments) != SEQUENCE_LENGTH:
                print(f"Warning: Expected {SEQUENCE_LENGTH} audio segments, got {len(audio_segments)}")
                # Pad or truncate to SEQUENCE_LENGTH
                while len(audio_segments) < SEQUENCE_LENGTH:
                    if audio_segments:
                        audio_segments.append(audio_segments[-1])
                    else:
                        # Calculate expected time dimension for mel spectrogram
                        expected_time_dim = int(AUDIO_SAMPLE_RATE * AUDIO_SEGMENT_LENGTH / 512) + 1
                        audio_segments.append(torch.zeros(AUDIO_N_MELS, expected_time_dim))
                audio_segments = audio_segments[:SEQUENCE_LENGTH]
            
            audio_features = torch.stack(audio_segments)  # [SEQUENCE_LENGTH, n_mels, time]
            
            return audio_features
            
        except Exception as e:
            print(f"Error extracting audio from {video_path}: {e}")
            # Return zeros if audio extraction fails - calculate expected time dimension
            expected_time_dim = int(AUDIO_SAMPLE_RATE * AUDIO_SEGMENT_LENGTH / 512) + 1
            return torch.zeros(SEQUENCE_LENGTH, AUDIO_N_MELS, expected_time_dim)
    
    def extract_video_features(self, video_path):
        """Extract video frames with temporal consistency"""
        # If preprocessing is enabled, this won't be called during training
        return self._extract_video_features(video_path)
    
    def _extract_video_features(self, video_path):
        """Internal method for video feature extraction"""
        try:
            cap = cv2.VideoCapture(str(video_path))
            total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
            fps = cap.get(cv2.CAP_PROP_FPS)
            
            if total_frames < SEQUENCE_LENGTH:
                print(f"Warning: Video {video_path} has only {total_frames} frames, need {SEQUENCE_LENGTH}")
            
            # Calculate frame indices to sample
            if total_frames >= SEQUENCE_LENGTH:
                frame_indices = np.linspace(0, total_frames - 1, SEQUENCE_LENGTH, dtype=int)
            else:
                # Repeat frames if video is too short
                frame_indices = np.linspace(0, total_frames - 1, total_frames, dtype=int)
                frame_indices = np.tile(frame_indices, (SEQUENCE_LENGTH // total_frames + 1))[:SEQUENCE_LENGTH]
            
            frames = []
            for frame_idx in frame_indices:
                cap.set(cv2.CAP_PROP_POS_FRAMES, frame_idx)
                ret, frame = cap.read()
                
                if ret:
                    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
                    frame = self.video_transform(frame)
                    frames.append(frame)
                else:
                    # Use last valid frame if read fails
                    if frames:
                        frames.append(frames[-1])
                    else:
                        frames.append(torch.zeros(3, IMG_SIZE, IMG_SIZE))
            
            cap.release()
            video_features = torch.stack(frames)  # [SEQUENCE_LENGTH, 3, IMG_SIZE, IMG_SIZE]
            
            return video_features
            
        except Exception as e:
            print(f"Error extracting video frames from {video_path}: {e}")
            return torch.zeros(SEQUENCE_LENGTH, 3, IMG_SIZE, IMG_SIZE)
    
    def __len__(self):
        return len(self.video_paths)
    
    def __getitem__(self, idx):
        video_path = self.video_paths[idx]
        label = self.labels[idx]
        
        # Use cached features if available, otherwise extract on-the-fly
        if self.preprocess and idx in self.audio_cache and idx in self.video_cache:
            audio_features = self.audio_cache[idx]
            video_features = self.video_cache[idx]
        else:
            # Extract features on-the-fly (slower)
            audio_features = self._extract_audio_features(video_path)
            video_features = self._extract_video_features(video_path)
        
        return {
            'audio': audio_features,
            'video': video_features,
            'label': torch.tensor(label, dtype=torch.long),
            'path': str(video_path)
        }

def load_dataset():
    """Load and split dataset"""
    print("\n" + "=" * 80)
    print("Section 2: Data Loading - STARTED")
    print("=" * 80)
    
    # Load metadata
    metadata_path = "/kaggle/input/hav-df-dataset/HAV-DF/video_metadata.csv"
    try:
        df = pd.read_csv(metadata_path)
        print(f"Loaded metadata with {len(df)} entries")
    except FileNotFoundError:
        print(f"Metadata file {metadata_path} not found. Creating sample metadata...")
        # Create sample metadata for demonstration
        sample_data = {
            'video_name': [f'sample_{i}.mp4' for i in range(10)],
            'label': ['REAL'] * 5 + ['FAKE'] * 5
        }
        df = pd.DataFrame(sample_data)
        df.to_csv(metadata_path, index=False)
        print(f"Created sample metadata with {len(df)} entries")
    except Exception as e:
        print(f"Error loading metadata: {e}")
        # Create minimal fallback metadata
        df = pd.DataFrame({'video_name': [], 'label': []})
        print("Using empty metadata")
    
    # Prepare data structures
    train_videos = []
    test_videos = []
    train_labels = []
    test_labels = []
    
    # Directories
    train_dir = "/kaggle/input/hav-df-dataset/HAV-DF/train_videos"
    test_dir = "/kaggle/input/hav-df-dataset/HAV-DF/test_videos"
    
    print(f"Looking for videos in:")
    print(f"  Train directory: {train_dir}")
    print(f"  Test directory: {test_dir}")
    print(f"  Current working directory: {os.getcwd()}")
    
    # Check if directories exist and list their contents
    if os.path.exists(train_dir):
        train_contents = os.listdir(train_dir)
        print(f"  Train directory contents: {len(train_contents)} items")
        video_files = [f for f in train_contents if f.endswith('.mp4')]
        print(f"  Train video files found: {len(video_files)}")
    else:
        print(f"  Train directory does not exist: {train_dir}")
    
    if os.path.exists(test_dir):
        test_contents = os.listdir(test_dir)
        print(f"  Test directory contents: {len(test_contents)} items")
        video_files = [f for f in test_contents if f.endswith('.mp4')]
        print(f"  Test video files found: {len(video_files)}")
    else:
        print(f"  Test directory does not exist: {test_dir}")
    
    # Process training videos
    if os.path.exists(train_dir):
        print("Processing training videos...")
        for filename in os.listdir(train_dir):
            if filename.endswith(".mp4"):
                video_file = os.path.join(train_dir, filename)
                video_name = filename
                if video_name in df['video_name'].values:
                    label = df[df['video_name'] == video_name]['label'].iloc[0]
                    train_videos.append(Path(video_file))
                    train_labels.append(0 if label == 'REAL' else 1)
                    # print(f"  Added: {filename} -> {label}")
        print(f"Found {len(train_videos)} training videos")
    else:
        print(f"Train directory {train_dir} does not exist")
    
    # Process test videos
    if os.path.exists(test_dir):
        print("Processing test videos...")
        for filename in os.listdir(test_dir):
            if filename.endswith(".mp4"):
                video_file = os.path.join(test_dir, filename)
                video_name = filename
                if video_name in df['video_name'].values:
                    label = df[df['video_name'] == video_name]['label'].iloc[0]
                    test_videos.append(Path(video_file))
                    test_labels.append(0 if label == 'REAL' else 1)
                    # print(f"  Added: {filename} -> {label}")
        print(f"Found {len(test_videos)} test videos")
    else:
        print(f"Test directory {test_dir} does not exist")
    
    print(f"\nDataset statistics:")
    print(f"  Train - Real: {sum(1 for l in train_labels if l == 0)}, Fake: {sum(1 for l in train_labels if l == 1)}")
    print(f"  Test - Real: {sum(1 for l in test_labels if l == 0)}, Fake: {sum(1 for l in test_labels if l == 1)}")
    
    # Create datasets
    if train_videos:
        # Split training data for validation
        train_size = int(len(train_videos) * (1 - VALIDATION_SPLIT))
        val_size = len(train_videos) - train_size
        
        indices = list(range(len(train_videos)))
        random.shuffle(indices)
        
        train_indices = indices[:train_size]
        val_indices = indices[train_size:]
        
        train_videos_split = [train_videos[i] for i in train_indices]
        train_labels_split = [train_labels[i] for i in train_indices]
        val_videos = [train_videos[i] for i in val_indices]
        val_labels = [train_labels[i] for i in val_indices]
        
        train_dataset = MultiModalDeepfakeDataset(train_videos_split, train_labels_split, mode='train', preprocess=USE_PREPROCESSING)
        val_dataset = MultiModalDeepfakeDataset(val_videos, val_labels, mode='val', preprocess=USE_PREPROCESSING)
        
        print(f"\nDataset split:")
        print(f"  Training: {len(train_dataset)} videos")
        print(f"  Validation: {len(val_dataset)} videos")
    else:
        train_dataset = val_dataset = None
        print("No training videos found!")
    
    test_dataset = None
    if test_videos:
        test_dataset = MultiModalDeepfakeDataset(test_videos, test_labels, mode='test', preprocess=USE_PREPROCESSING)
        print(f"  Test: {len(test_dataset)} videos")
    
    print("Section 2: Data Loading - COMPLETED ✓")
    print("=" * 80)
    
    return train_dataset, val_dataset, test_dataset 

# ===================================================================
# SECTION 3: AUDIO FEATURE EXTRACTION NETWORK
# ===================================================================

class AudioFeatureExtractor(nn.Module):
    """
    Simplified Audio Feature Extractor - REDUCED COMPLEXITY
    Focuses on temporal inconsistencies in audio patterns
    """
    def __init__(self, input_dim=AUDIO_N_MELS, hidden_dim=HIDDEN_DIM, num_layers=2):  # REDUCED layers
        super(AudioFeatureExtractor, self).__init__()
        
        print("Initializing Audio Feature Extractor...")
        
        # SIMPLIFIED Convolutional layers - REDUCED COMPLEXITY
        self.conv_layers = nn.Sequential(
            # First conv block
            nn.Conv2d(1, 16, kernel_size=(3, 3), padding=1),  # REDUCED channels
            nn.BatchNorm2d(16),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2, 2),
            nn.Dropout2d(DROPOUT_RATE),  # INCREASED dropout
            
            # Second conv block
            nn.Conv2d(16, 32, kernel_size=(3, 3), padding=1),  # REDUCED channels
            nn.BatchNorm2d(32),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2, 2),
            nn.Dropout2d(DROPOUT_RATE),  # INCREASED dropout
            
            # Third conv block - SIMPLIFIED
            nn.Conv2d(32, 64, kernel_size=(3, 3), padding=1),  # REDUCED channels
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            nn.AdaptiveAvgPool2d((2, 2))  # SMALLER adaptive pooling
        )
        
        # SIMPLIFIED Temporal modeling with transformer
        self.temporal_dim = 64 * 2 * 2  # After conv layers - REDUCED
        self.temporal_projection = nn.Linear(self.temporal_dim, hidden_dim)
        
        # Positional encoding for temporal information
        self.positional_encoding = self._create_positional_encoding(SEQUENCE_LENGTH, hidden_dim)
        
        # SIMPLIFIED Transformer encoder for temporal dependencies
        encoder_layer = nn.TransformerEncoderLayer(
            d_model=hidden_dim,
            nhead=4,  # REDUCED attention heads
            dim_feedforward=hidden_dim * 2,  # REDUCED feedforward dim
            dropout=DROPOUT_RATE,  # INCREASED dropout
            activation='gelu',
            batch_first=True
        )
        self.transformer_encoder = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)
        
        # SIMPLIFIED Output projection
        self.output_projection = nn.Sequential(
            nn.LayerNorm(hidden_dim),
            nn.Dropout(DROPOUT_RATE),  # INCREASED dropout
            nn.Linear(hidden_dim, hidden_dim)
        )
        
        print(f"Audio Feature Extractor initialized with {hidden_dim} hidden dimensions")
        
    def _create_positional_encoding(self, seq_len, d_model):
        """Create positional encoding for temporal information"""
        pe = torch.zeros(seq_len, d_model)
        position = torch.arange(0, seq_len, dtype=torch.float).unsqueeze(1)
        div_term = torch.exp(torch.arange(0, d_model, 2).float() * (-math.log(10000.0) / d_model))
        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)
        return pe.unsqueeze(0)  # [1, seq_len, d_model]
        
    def forward(self, audio_input):
        """
        Forward pass for audio feature extraction
        Input: [batch_size, sequence_length, n_mels, time_frames]
        Output: [batch_size, sequence_length, hidden_dim]
        """
        batch_size, seq_len, n_mels, time_frames = audio_input.shape
        
        # Ensure seq_len matches SEQUENCE_LENGTH
        if seq_len != SEQUENCE_LENGTH:
            print(f"Warning: Expected sequence length {SEQUENCE_LENGTH}, got {seq_len}")
            # Adjust if needed
            if seq_len > SEQUENCE_LENGTH:
                audio_input = audio_input[:, :SEQUENCE_LENGTH, :, :]
                seq_len = SEQUENCE_LENGTH
            else:
                # Pad if too short
                padding_needed = SEQUENCE_LENGTH - seq_len
                pad_tensor = audio_input[:, -1:, :, :].repeat(1, padding_needed, 1, 1)
                audio_input = torch.cat([audio_input, pad_tensor], dim=1)
                seq_len = SEQUENCE_LENGTH
        
        # Reshape for processing through conv layers
        # [batch_size * seq_len, 1, n_mels, time_frames]
        audio_reshaped = audio_input.reshape(-1, 1, n_mels, time_frames)
        
        # Extract spatial features
        conv_features = self.conv_layers(audio_reshaped)  # [batch_size * seq_len, 64, 2, 2]
        
        # Flatten spatial dimensions
        conv_features = conv_features.reshape(-1, self.temporal_dim)  # [batch_size * seq_len, temporal_dim]
        
        # Project to hidden dimension
        temporal_features = self.temporal_projection(conv_features)  # [batch_size * seq_len, hidden_dim]
        
        # Reshape back to sequence format
        temporal_features = temporal_features.reshape(batch_size, seq_len, -1)  # [batch_size, seq_len, hidden_dim]
        
        # Add positional encoding - ensure dimensions match
        pos_encoding = self.positional_encoding[:, :seq_len, :].to(audio_input.device)
        if temporal_features.shape[1] != pos_encoding.shape[1]:
            print(f"Shape mismatch: temporal_features {temporal_features.shape}, pos_encoding {pos_encoding.shape}")
            # Take only the required sequence length
            temporal_features = temporal_features[:, :SEQUENCE_LENGTH, :]
            pos_encoding = self.positional_encoding[:, :SEQUENCE_LENGTH, :].to(audio_input.device)
        
        temporal_features = temporal_features + pos_encoding
        
        # Apply transformer encoder for temporal modeling
        transformer_output = self.transformer_encoder(temporal_features)  # [batch_size, seq_len, hidden_dim]
        
        # Apply output projection
        output_features = self.output_projection(transformer_output)  # [batch_size, seq_len, hidden_dim]
        
        return output_features

class AudioTemporalConsistencyModule(nn.Module):
    """
    Novel module to detect temporal inconsistencies in audio
    Uses contrastive learning approach
    """
    def __init__(self, feature_dim=256):
        super(AudioTemporalConsistencyModule, self).__init__()
        
        self.feature_dim = feature_dim
        
        # Temporal difference encoder
        self.temporal_diff_encoder = nn.Sequential(
            nn.Linear(feature_dim * 2, feature_dim),
            nn.LayerNorm(feature_dim),
            nn.GELU(),
            nn.Dropout(0.2),
            nn.Linear(feature_dim, feature_dim // 2),
            nn.LayerNorm(feature_dim // 2),
            nn.GELU(),
            nn.Linear(feature_dim // 2, 1)  # Consistency score
        )
        
        print("Audio Temporal Consistency Module initialized")
        
    def forward(self, audio_features):
        """
        Compute temporal consistency scores
        Input: [batch_size, seq_len, feature_dim]
        Output: [batch_size, seq_len-1] consistency scores
        """
        batch_size, seq_len, feature_dim = audio_features.shape
        
        consistency_scores = []
        
        # Compare adjacent frames
        for i in range(seq_len - 1):
            current_frame = audio_features[:, i, :]  # [batch_size, feature_dim]
            next_frame = audio_features[:, i + 1, :]  # [batch_size, feature_dim]
            
            # Concatenate for difference encoding
            frame_pair = torch.cat([current_frame, next_frame], dim=1)  # [batch_size, feature_dim * 2]
            
            # Compute consistency score
            consistency = torch.sigmoid(self.temporal_diff_encoder(frame_pair))  # [batch_size, 1]
            consistency_scores.append(consistency)
        
        consistency_scores = torch.cat(consistency_scores, dim=1)  # [batch_size, seq_len-1]
        
        return consistency_scores

print("\nSection 3: Audio Feature Extraction Network - COMPLETED ✓") 

# ===================================================================
# SECTION 4: VIDEO FEATURE EXTRACTION NETWORK
# ===================================================================

class VideoFeatureExtractor(nn.Module):
    """
    SIMPLIFIED Video Feature Extractor - REDUCED COMPLEXITY
    Focuses on spatial-temporal inconsistencies in video frames
    """
    def __init__(self, hidden_dim=HIDDEN_DIM, num_layers=2):  # REDUCED layers
        super(VideoFeatureExtractor, self).__init__()
        
        print("Initializing Video Feature Extractor...")
        
        # Use SMALLER pretrained ResNet as backbone - REDUCED COMPLEXITY
        resnet = torchvision.models.resnet18(pretrained=True)  # CHANGED to ResNet18
        
        # Remove the final classification layer and avgpool
        self.backbone = nn.Sequential(*list(resnet.children())[:-2])
        
        # SIMPLIFIED Spatial attention mechanism
        self.spatial_attention = SpatialAttentionModule(512)  # REDUCED - ResNet18 has 512 channels
        
        # SMALLER Adaptive pooling to fixed size
        self.adaptive_pool = nn.AdaptiveAvgPool2d((2, 2))  # REDUCED pool size
        
        # SIMPLIFIED Temporal modeling
        self.spatial_dim = 512 * 2 * 2  # After backbone and pooling - REDUCED
        self.temporal_projection = nn.Linear(self.spatial_dim, hidden_dim)
        
        # Positional encoding for temporal information
        self.positional_encoding = self._create_positional_encoding(SEQUENCE_LENGTH, hidden_dim)
        
        # SIMPLIFIED Transformer encoder for temporal dependencies
        encoder_layer = nn.TransformerEncoderLayer(
            d_model=hidden_dim,
            nhead=4,  # REDUCED attention heads
            dim_feedforward=hidden_dim * 2,  # REDUCED feedforward dim
            dropout=DROPOUT_RATE,  # INCREASED dropout
            activation='gelu',
            batch_first=True
        )
        self.transformer_encoder = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)
        
        # SIMPLIFIED Output projection
        self.output_projection = nn.Sequential(
            nn.LayerNorm(hidden_dim),
            nn.Dropout(DROPOUT_RATE),  # INCREASED dropout
            nn.Linear(hidden_dim, hidden_dim)
        )
        
        print(f"Video Feature Extractor initialized with {hidden_dim} hidden dimensions")
        
    def _create_positional_encoding(self, seq_len, d_model):
        """Create positional encoding for temporal information"""
        pe = torch.zeros(seq_len, d_model)
        position = torch.arange(0, seq_len, dtype=torch.float).unsqueeze(1)
        div_term = torch.exp(torch.arange(0, d_model, 2).float() * (-math.log(10000.0) / d_model))
        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)
        return pe.unsqueeze(0)  # [1, seq_len, d_model]
        
    def forward(self, video_input):
        """
        Forward pass for video feature extraction
        Input: [batch_size, sequence_length, 3, H, W]
        Output: [batch_size, sequence_length, hidden_dim]
        """
        batch_size, seq_len, channels, height, width = video_input.shape
        
        # Ensure seq_len matches SEQUENCE_LENGTH
        if seq_len != SEQUENCE_LENGTH:
            print(f"Warning: Video sequence length mismatch. Expected {SEQUENCE_LENGTH}, got {seq_len}. Adjusting.")
            if seq_len > SEQUENCE_LENGTH:
                video_input = video_input[:, :SEQUENCE_LENGTH, :, :, :]
                seq_len = SEQUENCE_LENGTH
            else:
                # Pad if too short
                padding_needed = SEQUENCE_LENGTH - seq_len
                pad_tensor = video_input[:, -1:, :, :, :].repeat(1, padding_needed, 1, 1, 1)
                video_input = torch.cat([video_input, pad_tensor], dim=1)
                seq_len = SEQUENCE_LENGTH
        
        # Reshape for processing through backbone
        video_reshaped = video_input.reshape(-1, channels, height, width)  # [batch_size * seq_len, 3, H, W]
        
        # Extract spatial features using ResNet backbone
        spatial_features = self.backbone(video_reshaped)  # [batch_size * seq_len, 2048, H', W']
        
        # Apply spatial attention
        attended_features = self.spatial_attention(spatial_features)  # [batch_size * seq_len, 2048, H', W']
        
        # Adaptive pooling to fixed size
        pooled_features = self.adaptive_pool(attended_features)  # [batch_size * seq_len, 2048, 4, 4]
        
        # Flatten spatial dimensions
        flattened_features = pooled_features.reshape(-1, self.spatial_dim)  # [batch_size * seq_len, spatial_dim]
        
        # Project to hidden dimension
        temporal_features = self.temporal_projection(flattened_features)  # [batch_size * seq_len, hidden_dim]
        
        # Reshape back to sequence format
        temporal_features = temporal_features.reshape(batch_size, seq_len, -1)  # [batch_size, seq_len, hidden_dim]
        
        # Add positional encoding
        pos_encoding = self.positional_encoding[:, :seq_len, :].to(video_input.device)
        temporal_features = temporal_features + pos_encoding
        
        # Apply transformer encoder for temporal modeling
        transformer_output = self.transformer_encoder(temporal_features)  # [batch_size, seq_len, hidden_dim]
        
        # Apply output projection
        output_features = self.output_projection(transformer_output)  # [batch_size, seq_len, hidden_dim]
        
        return output_features

class SpatialAttentionModule(nn.Module):
    """
    Spatial attention mechanism to focus on important regions
    """
    def __init__(self, in_channels, reduction_ratio=16):
        super(SpatialAttentionModule, self).__init__()
        
        # Channel attention
        self.channel_attention = nn.Sequential(
            nn.AdaptiveAvgPool2d(1),
            nn.Conv2d(in_channels, in_channels // reduction_ratio, 1),
            nn.ReLU(inplace=True),
            nn.Conv2d(in_channels // reduction_ratio, in_channels, 1),
            nn.Sigmoid()
        )
        
        # Spatial attention
        self.spatial_attention = nn.Sequential(
            nn.Conv2d(2, 1, kernel_size=7, padding=3),
            nn.Sigmoid()
        )
        
    def forward(self, x):
        # Channel attention
        channel_att = self.channel_attention(x)
        x = x * channel_att
        
        # Spatial attention
        avg_pool = torch.mean(x, dim=1, keepdim=True)
        max_pool, _ = torch.max(x, dim=1, keepdim=True)
        spatial_input = torch.cat([avg_pool, max_pool], dim=1)
        spatial_att = self.spatial_attention(spatial_input)
        x = x * spatial_att
        
        return x

class VideoTemporalConsistencyModule(nn.Module):
    """
    Novel module to detect temporal inconsistencies in video
    Uses optical flow and frame difference analysis
    """
    def __init__(self, feature_dim=256):
        super(VideoTemporalConsistencyModule, self).__init__()
        
        self.feature_dim = feature_dim
        
        # Frame difference encoder
        self.frame_diff_encoder = nn.Sequential(
            nn.Linear(feature_dim * 2, feature_dim),
            nn.LayerNorm(feature_dim),
            nn.GELU(),
            nn.Dropout(0.2),
            nn.Linear(feature_dim, feature_dim // 2),
            nn.LayerNorm(feature_dim // 2),
            nn.GELU(),
            nn.Linear(feature_dim // 2, 1)  # Consistency score
        )
        
        # Motion consistency encoder
        self.motion_encoder = nn.Sequential(
            nn.Linear(feature_dim * 3, feature_dim),
            nn.LayerNorm(feature_dim),
            nn.GELU(),
            nn.Dropout(0.2),
            nn.Linear(feature_dim, feature_dim // 2),
            nn.LayerNorm(feature_dim // 2),
            nn.GELU(),
            nn.Linear(feature_dim // 2, 1)  # Motion consistency score
        )
        
        print("Video Temporal Consistency Module initialized")
        
    def forward(self, video_features):
        """
        Compute temporal consistency scores for video
        Input: [batch_size, seq_len, feature_dim]
        Output: [batch_size, seq_len-1] consistency scores
        """
        batch_size, seq_len, feature_dim = video_features.shape
        
        frame_consistency_scores = []
        motion_consistency_scores = []
        
        # Compare adjacent frames
        for i in range(seq_len - 1):
            current_frame = video_features[:, i, :]  # [batch_size, feature_dim]
            next_frame = video_features[:, i + 1, :]  # [batch_size, feature_dim]
            
            # Frame difference consistency
            frame_pair = torch.cat([current_frame, next_frame], dim=1)  # [batch_size, feature_dim * 2]
            frame_consistency = torch.sigmoid(self.frame_diff_encoder(frame_pair))  # [batch_size, 1]
            frame_consistency_scores.append(frame_consistency)
            
            # Motion consistency (using triplet of frames)
            if i < seq_len - 2:
                third_frame = video_features[:, i + 2, :]  # [batch_size, feature_dim]
                motion_triplet = torch.cat([current_frame, next_frame, third_frame], dim=1)
                motion_consistency = torch.sigmoid(self.motion_encoder(motion_triplet))
                motion_consistency_scores.append(motion_consistency)
        
        frame_consistency_scores = torch.cat(frame_consistency_scores, dim=1)  # [batch_size, seq_len-1]
        
        if motion_consistency_scores:
            motion_consistency_scores = torch.cat(motion_consistency_scores, dim=1)  # [batch_size, seq_len-2]
            # Pad motion scores to match frame consistency length
            motion_consistency_scores = F.pad(motion_consistency_scores, (0, 1), value=0.5)
        else:
            motion_consistency_scores = torch.full_like(frame_consistency_scores, 0.5)
        
        # Combine frame and motion consistency
        combined_consistency = (frame_consistency_scores + motion_consistency_scores) / 2
        
        return combined_consistency

print("Section 4: Video Feature Extraction Network - COMPLETED ✓") 

# ===================================================================
# SECTION 5: CROSS-MODAL FUSION NETWORK
# ===================================================================

class CrossModalAttention(nn.Module):
    """
    Novel Cross-Modal Attention mechanism between audio and video features
    Uses multi-head attention to learn cross-modal relationships
    """
    def __init__(self, feature_dim=256, num_heads=8, dropout=0.1):
        super(CrossModalAttention, self).__init__()
        
        self.feature_dim = feature_dim
        self.num_heads = num_heads
        self.head_dim = feature_dim // num_heads
        
        assert feature_dim % num_heads == 0, "feature_dim must be divisible by num_heads"
        
        # Linear projections for Q, K, V
        self.query_proj = nn.Linear(feature_dim, feature_dim)
        self.key_proj = nn.Linear(feature_dim, feature_dim)
        self.value_proj = nn.Linear(feature_dim, feature_dim)
        
        # Output projection
        self.output_proj = nn.Linear(feature_dim, feature_dim)
        self.dropout = nn.Dropout(dropout)
        
        # Layer normalization
        self.layer_norm = nn.LayerNorm(feature_dim)
        
        print(f"Cross-Modal Attention initialized with {num_heads} heads")
        
    def forward(self, query_features, key_value_features):
        """
        Cross-modal attention between query and key-value features
        Input:
            query_features: [batch_size, seq_len, feature_dim]
            key_value_features: [batch_size, seq_len, feature_dim]
        Output:
            attended_features: [batch_size, seq_len, feature_dim]
        """
        batch_size, seq_len, feature_dim = query_features.shape
        
        # Linear projections
        Q = self.query_proj(query_features)  # [batch_size, seq_len, feature_dim]
        K = self.key_proj(key_value_features)  # [batch_size, seq_len, feature_dim]
        V = self.value_proj(key_value_features)  # [batch_size, seq_len, feature_dim]
        
        # Reshape for multi-head attention
        Q = Q.reshape(batch_size, seq_len, self.num_heads, self.head_dim).transpose(1, 2)
        K = K.reshape(batch_size, seq_len, self.num_heads, self.head_dim).transpose(1, 2)
        V = V.reshape(batch_size, seq_len, self.num_heads, self.head_dim).transpose(1, 2)
        
        # Compute attention scores
        attention_scores = torch.matmul(Q, K.transpose(-2, -1)) / math.sqrt(self.head_dim)
        attention_weights = F.softmax(attention_scores, dim=-1)
        attention_weights = self.dropout(attention_weights)
        
        # Apply attention to values
        attended_values = torch.matmul(attention_weights, V)
        
        # Reshape back to original format
        attended_values = attended_values.transpose(1, 2).contiguous().reshape(
            batch_size, seq_len, feature_dim
        )
        
        # Output projection and residual connection
        output = self.output_proj(attended_values)
        output = self.layer_norm(output + query_features)
        
        return output, attention_weights

class MultiModalFusionNetwork(nn.Module):
    """
    SIMPLIFIED Multi-Modal Fusion Network - REDUCED COMPLEXITY
    """
    def __init__(self, feature_dim=HIDDEN_DIM, num_classes=2, num_fusion_layers=1):  # REDUCED layers
        super(MultiModalFusionNetwork, self).__init__()
        
        self.feature_dim = feature_dim
        self.num_classes = num_classes
        self.num_fusion_layers = num_fusion_layers
        
        print("Initializing Multi-Modal Fusion Network...")
        
        # SIMPLIFIED Cross-modal attention layers
        self.audio_to_video_attention = nn.ModuleList([
            CrossModalAttention(feature_dim, num_heads=4, dropout=DROPOUT_RATE) for _ in range(num_fusion_layers)  # REDUCED heads
        ])
        self.video_to_audio_attention = nn.ModuleList([
            CrossModalAttention(feature_dim, num_heads=4, dropout=DROPOUT_RATE) for _ in range(num_fusion_layers)  # REDUCED heads
        ])
        
        # SIMPLIFIED Self-attention for refined features
        self.self_attention = nn.ModuleList([
            nn.MultiheadAttention(feature_dim, num_heads=4, dropout=DROPOUT_RATE, batch_first=True)  # REDUCED heads
            for _ in range(2)
        ])
        
        # SIMPLIFIED Feature fusion strategies - REDUCED to key ones only
        self.fusion_strategies = nn.ModuleDict({
            'concat': nn.Sequential(
                nn.Linear(feature_dim * 2, feature_dim),
                nn.LayerNorm(feature_dim),
                nn.GELU(),
                nn.Dropout(DROPOUT_RATE)  # INCREASED dropout
            ),
            'attention': nn.Sequential(
                nn.Linear(feature_dim * 2, feature_dim),
                nn.Tanh(),
                nn.Dropout(DROPOUT_RATE),  # Added dropout
                nn.Linear(feature_dim, 1),
                nn.Softmax(dim=1)
            )
        })
        
        # Temporal aggregation
        self.temporal_aggregator = TemporalAggregator(feature_dim)
        
        # SIMPLIFIED Final classification layers
        self.classifier = nn.Sequential(
            nn.Linear(feature_dim * 2, feature_dim),  # REDUCED - only 2 fusion strategies now
            nn.LayerNorm(feature_dim),
            nn.GELU(),
            nn.Dropout(DROPOUT_RATE),  # INCREASED dropout
            nn.Linear(feature_dim, num_classes)
        )
        
        # Consistency prediction heads
        self.audio_consistency_head = nn.Linear(feature_dim, 1)
        self.video_consistency_head = nn.Linear(feature_dim, 1)
        self.cross_modal_consistency_head = nn.Linear(feature_dim * 2, 1)
        
        print(f"Multi-Modal Fusion Network initialized with {feature_dim} feature dimensions")
        
    def forward(self, audio_features, video_features):
        """
        Forward pass for multi-modal fusion
        Input:
            audio_features: [batch_size, seq_len, feature_dim]
            video_features: [batch_size, seq_len, feature_dim]
        Output:
            logits: [batch_size, num_classes]
            consistency_scores: dict of consistency predictions
        """
        batch_size, seq_len, feature_dim = audio_features.shape
        
        # print(f"Processing batch: {batch_size} samples, sequence length: {seq_len}")
        
        # Store original features for residual connections
        orig_audio_features = audio_features.clone()
        orig_video_features = video_features.clone()
        
        # Multi-layer cross-modal attention
        for i in range(self.num_fusion_layers):
            # Audio attended by video
            audio_attended, audio_attention_weights = self.audio_to_video_attention[i](
                audio_features, video_features
            )
            
            # Video attended by audio
            video_attended, video_attention_weights = self.video_to_audio_attention[i](
                video_features, audio_features
            )
            
            # Update features
            audio_features = audio_attended
            video_features = video_attended
            
            # print(f"Fusion layer {i+1}: Cross-modal attention applied")
        
        # Self-attention for feature refinement
        audio_refined, _ = self.self_attention[0](audio_features, audio_features, audio_features)
        video_refined, _ = self.self_attention[1](video_features, video_features, video_features)
        
        # print("Self-attention refinement completed")
        
        # SIMPLIFIED fusion strategies - only key ones
        fused_features = []
        
        # 1. Concatenation fusion
        concat_fusion = torch.cat([audio_refined, video_refined], dim=-1)  # [batch_size, seq_len, 2*feature_dim]
        concat_fusion = self.fusion_strategies['concat'](concat_fusion)  # [batch_size, seq_len, feature_dim]
        fused_features.append(concat_fusion)
        
        # 2. Attention-based fusion
        attention_input = torch.cat([audio_refined, video_refined], dim=-1)
        attention_weights = self.fusion_strategies['attention'](attention_input)  # [batch_size, seq_len, 1]
        attention_fusion = attention_weights * audio_refined + (1 - attention_weights) * video_refined
        fused_features.append(attention_fusion)
        
        # print("Multiple fusion strategies applied")
        
        # Temporal aggregation for each fusion strategy
        aggregated_features = []
        for i, fusion_feat in enumerate(fused_features):
            aggregated = self.temporal_aggregator(fusion_feat)  # [batch_size, feature_dim]
            aggregated_features.append(aggregated)
            # print(f"Temporal aggregation {i+1} completed")
        
        # Combine all fusion strategies
        final_features = torch.cat(aggregated_features, dim=-1)  # [batch_size, 2*feature_dim] - REDUCED
        
        # Classification
        logits = self.classifier(final_features)  # [batch_size, num_classes]
        
        # Consistency predictions
        audio_global = self.temporal_aggregator(audio_refined)  # [batch_size, feature_dim]
        video_global = self.temporal_aggregator(video_refined)  # [batch_size, feature_dim]
        
        consistency_scores = {
            'audio_consistency': torch.sigmoid(self.audio_consistency_head(audio_global)),
            'video_consistency': torch.sigmoid(self.video_consistency_head(video_global)),
            'cross_modal_consistency': torch.sigmoid(
                self.cross_modal_consistency_head(torch.cat([audio_global, video_global], dim=-1))
            )
        }
        
        # print("Classification and consistency prediction completed")
        
        return logits, consistency_scores

class TemporalAggregator(nn.Module):
    """
    Temporal aggregation module using multiple pooling strategies
    """
    def __init__(self, feature_dim):
        super(TemporalAggregator, self).__init__()
        
        self.feature_dim = feature_dim
        
        # Attention-based aggregation
        self.attention_aggregator = nn.Sequential(
            nn.Linear(feature_dim, feature_dim // 2),
            nn.Tanh(),
            nn.Linear(feature_dim // 2, 1),
            nn.Softmax(dim=1)
        )
        
        # Learnable pooling weights
        self.pooling_weights = nn.Parameter(torch.ones(4) / 4)  # 4 pooling strategies
        
    def forward(self, features):
        """
        Aggregate temporal features using multiple strategies
        Input: [batch_size, seq_len, feature_dim]
        Output: [batch_size, feature_dim]
        """
        # 1. Attention-based aggregation
        attention_weights = self.attention_aggregator(features)  # [batch_size, seq_len, 1]
        attention_pooled = torch.sum(features * attention_weights, dim=1)  # [batch_size, feature_dim]
        
        # 2. Average pooling
        avg_pooled = torch.mean(features, dim=1)  # [batch_size, feature_dim]
        
        # 3. Max pooling
        max_pooled, _ = torch.max(features, dim=1)  # [batch_size, feature_dim]
        
        # 4. Last frame
        last_pooled = features[:, -1, :]  # [batch_size, feature_dim]
        
        # Combine with learnable weights
        pooled_features = [attention_pooled, avg_pooled, max_pooled, last_pooled]
        weighted_pooled = sum(w * feat for w, feat in zip(self.pooling_weights, pooled_features))
        
        return weighted_pooled

print("Section 5: Cross-Modal Fusion Network - COMPLETED ✓") 

# ===================================================================
# SECTION 6: COMPLETE MULTIMODAL MODEL AND TRAINING
# ===================================================================

class MultiModalDeepfakeDetector(nn.Module):
    """
    OPTIMIZED Multi-Modal Deepfake Detection System - REDUCED COMPLEXITY
    Architecture optimized for small datasets with strong regularization
    """
    def __init__(self, hidden_dim=HIDDEN_DIM, num_classes=2):
        super(MultiModalDeepfakeDetector, self).__init__()
        
        print("\n" + "=" * 80)
        print("INITIALIZING COMPLETE MULTIMODAL DEEPFAKE DETECTOR")
        print("=" * 80)
        
        self.hidden_dim = hidden_dim
        self.num_classes = num_classes
        
        # Audio processing pipeline
        self.audio_extractor = AudioFeatureExtractor(hidden_dim=hidden_dim)
        self.audio_consistency = AudioTemporalConsistencyModule(hidden_dim)
        
        # Video processing pipeline
        self.video_extractor = VideoFeatureExtractor(hidden_dim=hidden_dim)
        self.video_consistency = VideoTemporalConsistencyModule(hidden_dim)
        
        # Multi-modal fusion
        self.fusion_network = MultiModalFusionNetwork(
            feature_dim=hidden_dim, 
            num_classes=num_classes
        )
        
        print("Complete Multi-Modal Deepfake Detector initialized successfully!")
        print("=" * 80)
        
    def forward(self, audio_input, video_input, return_features=False):
        """
        Forward pass through the complete model
        """
        # Extract audio features
        audio_features = self.audio_extractor(audio_input)  # [batch_size, seq_len, hidden_dim]
        
        # Extract video features
        video_features = self.video_extractor(video_input)  # [batch_size, seq_len, hidden_dim]
        
        # Compute temporal consistency scores
        audio_temp_consistency = self.audio_consistency(audio_features)  # [batch_size, seq_len-1]
        video_temp_consistency = self.video_consistency(video_features)  # [batch_size, seq_len-1]
        
        # Fusion and classification
        logits, consistency_scores = self.fusion_network(audio_features, video_features)
        
        outputs = {
            'logits': logits,
            'consistency_scores': consistency_scores,
            'audio_temporal_consistency': audio_temp_consistency,
            'video_temporal_consistency': video_temp_consistency
        }
        
        if return_features:
            outputs.update({
                'audio_features': audio_features,
                'video_features': video_features
            })
        
        return outputs

class MultiModalLoss(nn.Module):
    """
    Enhanced Multi-Modal Loss with DYNAMIC CLASS BALANCING and LABEL SMOOTHING
    """
    def __init__(self, class_counts=None, alpha=1.0, beta=0.3, gamma=0.1):  # REDUCED consistency weights
        super(MultiModalLoss, self).__init__()
        
        self.alpha = alpha  # Classification loss weight
        self.beta = beta    # Consistency loss weight (REDUCED)
        self.gamma = gamma  # Contrastive loss weight (REDUCED)
        
        # DYNAMIC CLASS BALANCING
        if class_counts is not None:
            total_samples = sum(class_counts)
            class_weights = [total_samples / (len(class_counts) * count) for count in class_counts]
            self.class_weights = torch.tensor(class_weights, dtype=torch.float32)
            print(f"Class weights calculated: {self.class_weights}")
        else:
            self.class_weights = None
        
        # LABEL SMOOTHING for regularization
        self.classification_loss = nn.CrossEntropyLoss(
            weight=self.class_weights,
            label_smoothing=LABEL_SMOOTHING
        )
        self.consistency_loss = nn.BCELoss()
        
        print(f"Enhanced Multi-Modal Loss initialized:")
        print(f"  Weights: α={alpha}, β={beta}, γ={gamma}")
        print(f"  Label smoothing: {LABEL_SMOOTHING}")
        print(f"  Dynamic class balancing: {'Enabled' if class_counts else 'Disabled'}")
        
    def forward(self, outputs, labels):
        """
        Compute multi-modal loss
        """
        logits = outputs['logits']
        consistency_scores = outputs['consistency_scores']
        audio_temp_consistency = outputs['audio_temporal_consistency']
        video_temp_consistency = outputs['video_temporal_consistency']
        
        # 1. Classification loss
        cls_loss = self.classification_loss(logits, labels)
        
        # 2. Consistency losses
        # For real videos, consistency should be high; for fake videos, consistency should be low
        real_mask = (labels == 0).float()
        fake_mask = (labels == 1).float()
        
        # Global consistency targets
        audio_consistency_target = real_mask.unsqueeze(1)  # [batch_size, 1]
        video_consistency_target = real_mask.unsqueeze(1)
        cross_modal_consistency_target = real_mask.unsqueeze(1)
        
        # Global consistency losses
        audio_consistency_loss = self.consistency_loss(
            consistency_scores['audio_consistency'], audio_consistency_target
        )
        video_consistency_loss = self.consistency_loss(
            consistency_scores['video_consistency'], video_consistency_target
        )
        cross_modal_consistency_loss = self.consistency_loss(
            consistency_scores['cross_modal_consistency'], cross_modal_consistency_target
        )
        
        # Temporal consistency losses
        # Real videos should have higher temporal consistency
        temporal_consistency_target = real_mask.unsqueeze(1).expand(-1, audio_temp_consistency.shape[1])
        
        audio_temporal_loss = self.consistency_loss(audio_temp_consistency, temporal_consistency_target)
        video_temporal_loss = self.consistency_loss(video_temp_consistency, temporal_consistency_target)
        
        # Total consistency loss
        consistency_loss = (
            audio_consistency_loss + video_consistency_loss + cross_modal_consistency_loss +
            audio_temporal_loss + video_temporal_loss
        ) / 5.0
        
        # 3. Contrastive loss (InfoNCE style)
        if 'audio_features' in outputs and 'video_features' in outputs:
            contrastive_loss = self.compute_contrastive_loss(
                outputs['audio_features'], outputs['video_features'], labels
            )
        else:
            contrastive_loss = torch.tensor(0.0, device=logits.device)
        
        # Total loss
        total_loss = (
            self.alpha * cls_loss +
            self.beta * consistency_loss +
            self.gamma * contrastive_loss
        )
        
        return {
            'total_loss': total_loss,
            'classification_loss': cls_loss,
            'consistency_loss': consistency_loss,
            'contrastive_loss': contrastive_loss
        }
    
    def compute_contrastive_loss(self, audio_features, video_features, labels):
        """
        Compute contrastive loss between audio and video features
        """
        batch_size, seq_len, feature_dim = audio_features.shape
        
        # Global average pooling
        audio_global = torch.mean(audio_features, dim=1)  # [batch_size, feature_dim]
        video_global = torch.mean(video_features, dim=1)  # [batch_size, feature_dim]
        
        # Normalize features
        audio_global = F.normalize(audio_global, dim=1)
        video_global = F.normalize(video_global, dim=1)
        
        # Compute similarity matrix
        similarity_matrix = torch.matmul(audio_global, video_global.T)  # [batch_size, batch_size]
        
        # Create positive mask (same sample)
        positive_mask = torch.eye(batch_size, device=audio_features.device)
        
        # InfoNCE loss
        temperature = 0.1
        exp_sim = torch.exp(similarity_matrix / temperature)
        
        # Positive similarities
        positive_sim = torch.diag(exp_sim)
        
        # All similarities for each sample
        all_sim = torch.sum(exp_sim, dim=1)
        
        # Contrastive loss
        contrastive_loss = -torch.log(positive_sim / all_sim).mean()
        
        return contrastive_loss

def train_model(model, train_loader, val_loader, num_epochs=NUM_EPOCHS):
    """
    ENHANCED Training function with STRONG REGULARIZATION and CLASS BALANCING
    """
    print("\n" + "=" * 80)
    print("STARTING ENHANCED TRAINING PROCESS")
    print("=" * 80)
    
    # Calculate class distribution for dynamic balancing
    train_labels = []
    for batch in train_loader:
        train_labels.extend(batch['label'].numpy().tolist())
    
    class_counts = [train_labels.count(0), train_labels.count(1)]  # [real_count, fake_count]
    print(f"Training class distribution: Real={class_counts[0]}, Fake={class_counts[1]}")
    
    # Initialize optimizer with STRONGER REGULARIZATION
    optimizer = optim.AdamW(
        model.parameters(),
        lr=LEARNING_RATE,
        weight_decay=WEIGHT_DECAY,  # INCREASED weight decay
        betas=(0.9, 0.999)
    )
    
    # More conservative scheduler for small dataset
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(
        optimizer, mode='max', factor=0.5, patience=5, verbose=True
    )
    
    # Initialize loss function with DYNAMIC CLASS BALANCING
    criterion = MultiModalLoss(class_counts=class_counts, alpha=1.0, beta=0.3, gamma=0.1).to(device)
    
    # Training history
    train_history = {
        'train_loss': [], 'train_acc': [],
        'val_loss': [], 'val_acc': [],
        'learning_rates': []
    }
    
    best_val_acc = 0.0
    patience = 10
    patience_counter = 0
    
    print(f"ENHANCED Training configuration:")
    print(f"  Optimizer: AdamW with lr={LEARNING_RATE}, weight_decay={WEIGHT_DECAY}")
    print(f"  Scheduler: ReduceLROnPlateau")
    print(f"  Loss: Enhanced Multi-Modal Loss with class balancing")
    print(f"  Dropout rate: {DROPOUT_RATE}")
    print(f"  Label smoothing: {LABEL_SMOOTHING}")
    print(f"  Early stopping patience: {patience}")
    print("=" * 80)
    
    for epoch in range(num_epochs):
        print(f"\nEPOCH {epoch + 1}/{num_epochs}")
        print("-" * 50)
        
        # Training phase
        model.train()
        train_losses = []
        train_predictions = []
        train_labels = []
        
        print("Training phase:")
        
        # Training progress bar
        train_pbar = tqdm(enumerate(train_loader), total=len(train_loader), 
                         desc=f"Epoch {epoch+1} Training", leave=False)
        
        for batch_idx, batch in train_pbar:
            batch_start_time = time.time()
            
            audio_data = batch['audio'].to(device)
            video_data = batch['video'].to(device)
            labels = batch['label'].to(device)
            
            # Forward pass
            optimizer.zero_grad()
            outputs = model(audio_data, video_data, return_features=True)
            
            # Compute loss
            loss_dict = criterion(outputs, labels)
            total_loss = loss_dict['total_loss']
            
            # Backward pass
            total_loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            optimizer.step()
            
            # Record metrics
            train_losses.append(total_loss.item())
            predictions = torch.argmax(outputs['logits'], dim=1)
            train_predictions.extend(predictions.cpu().numpy())
            train_labels.extend(labels.cpu().numpy())
            
            # Update progress bar
            batch_time = time.time() - batch_start_time
            train_pbar.set_postfix({
                'Loss': f"{total_loss.item():.4f}",
                'Acc': f"{accuracy_score(train_labels, train_predictions):.3f}",
                'Time/batch': f"{batch_time:.2f}s"
            })
        
        train_pbar.close()
        
        # Calculate training metrics
        train_loss = np.mean(train_losses)
        train_acc = accuracy_score(train_labels, train_predictions)
        
        # Validation phase
        model.eval()
        val_losses = []
        val_predictions = []
        val_labels = []
        
        print("\nValidation phase:")
        
        # Validation progress bar
        val_pbar = tqdm(enumerate(val_loader), total=len(val_loader), 
                       desc=f"Epoch {epoch+1} Validation", leave=False)
        
        with torch.no_grad():
            for batch_idx, batch in val_pbar:
                audio_data = batch['audio'].to(device)
                video_data = batch['video'].to(device)
                labels = batch['label'].to(device)
                
                outputs = model(audio_data, video_data)
                loss_dict = criterion(outputs, labels)
                
                val_losses.append(loss_dict['total_loss'].item())
                predictions = torch.argmax(outputs['logits'], dim=1)
                val_predictions.extend(predictions.cpu().numpy())
                val_labels.extend(labels.cpu().numpy())
                
                # Update progress bar
                val_pbar.set_postfix({
                    'Loss': f"{loss_dict['total_loss'].item():.4f}",
                    'Acc': f"{accuracy_score(val_labels, val_predictions):.3f}"
                })
        
        val_pbar.close()
        
        # Calculate validation metrics
        val_loss = np.mean(val_losses)
        val_acc = accuracy_score(val_labels, val_predictions)
        
        # Update learning rate based on validation accuracy
        scheduler.step(val_acc)
        current_lr = optimizer.param_groups[0]['lr']
        
        # Record history
        train_history['train_loss'].append(train_loss)
        train_history['train_acc'].append(train_acc)
        train_history['val_loss'].append(val_loss)
        train_history['val_acc'].append(val_acc)
        train_history['learning_rates'].append(current_lr)
        
        # Print epoch summary
        print(f"\nEpoch {epoch + 1} Summary:")
        print(f"  Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.4f}")
        print(f"  Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.4f}")
        print(f"  Learning Rate: {current_lr:.6f}")
        
        # Early stopping and model saving
        if val_acc > best_val_acc:
            best_val_acc = val_acc
            patience_counter = 0
            torch.save({
                'epoch': epoch + 1,
                'model_state_dict': model.state_dict(),
                'optimizer_state_dict': optimizer.state_dict(),
                'best_val_acc': best_val_acc,
                'train_history': train_history
            }, 'best_model.pth')
            print(f"  ✓ New best validation accuracy: {best_val_acc:.4f}")
        else:
            patience_counter += 1
            print(f"  Patience: {patience_counter}/{patience}")
        
        if patience_counter >= patience:
            print(f"\nEarly stopping triggered after {epoch + 1} epochs")
            break
        
        print("=" * 80)
    
    print("\nTraining completed!")
    print(f"Best validation accuracy: {best_val_acc:.4f}")
    
    return train_history

print("Section 6: Complete Model and Training - COMPLETED ✓") 

# ===================================================================
# SECTION 7: EVALUATION AND ANALYSIS
# ===================================================================

def evaluate_model(model, test_loader, save_results=True):
    """
    Comprehensive model evaluation with detailed analysis
    """
    print("\n" + "=" * 80)
    print("STARTING MODEL EVALUATION")
    print("=" * 80)
    
    model.eval()
    criterion = MultiModalLoss().to(device)  # Use default weights for evaluation
    
    all_predictions = []
    all_labels = []
    all_probabilities = []
    test_losses = []
    consistency_analysis = defaultdict(list)
    
    print("Evaluating on test set...")
    
    # Evaluation progress bar
    eval_pbar = tqdm(enumerate(test_loader), total=len(test_loader), 
                    desc="Evaluating", leave=False)
    
    with torch.no_grad():
        for batch_idx, batch in eval_pbar:
            audio_data = batch['audio'].to(device)
            video_data = batch['video'].to(device)
            labels = batch['label'].to(device)
            paths = batch['path']
            
            # Forward pass
            outputs = model(audio_data, video_data, return_features=True)
            
            # Compute loss
            loss_dict = criterion(outputs, labels)
            test_losses.append(loss_dict['total_loss'].item())
            
            # Get predictions and probabilities
            logits = outputs['logits']
            probabilities = F.softmax(logits, dim=1)
            predictions = torch.argmax(logits, dim=1)
            
            # Store results
            all_predictions.extend(predictions.cpu().numpy().tolist())
            all_labels.extend(labels.cpu().numpy().tolist())
            all_probabilities.extend(probabilities.cpu().numpy().tolist())
            
            # Analyze consistency scores
            consistency_scores = outputs['consistency_scores']
            for i, (pred, label, path) in enumerate(zip(predictions, labels, paths)):
                consistency_analysis['predictions'].append(pred.item())
                consistency_analysis['labels'].append(label.item())
                consistency_analysis['paths'].append(path)
                consistency_analysis['audio_consistency'].append(
                    consistency_scores['audio_consistency'][i].item()
                )
                consistency_analysis['video_consistency'].append(
                    consistency_scores['video_consistency'][i].item()
                )
                consistency_analysis['cross_modal_consistency'].append(
                    consistency_scores['cross_modal_consistency'][i].item()
                )
            
            # Update progress bar
            eval_pbar.set_postfix({
                'Loss': f"{loss_dict['total_loss'].item():.4f}",
                'Acc': f"{accuracy_score(all_labels, all_predictions):.3f}"
            })
    
    eval_pbar.close()
    
    # Calculate metrics
    test_loss = np.mean(test_losses)
    test_accuracy = accuracy_score(all_labels, all_predictions)
    precision, recall, f1, _ = precision_recall_fscore_support(
        all_labels, all_predictions, average='weighted'
    )
    
    # Confusion matrix
    cm = confusion_matrix(all_labels, all_predictions)
    
    # Print results
    print("\n" + "=" * 80)
    print("EVALUATION RESULTS")
    print("=" * 80)
    print(f"Test Loss: {test_loss:.4f}")
    print(f"Test Accuracy: {test_accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1-Score: {f1:.4f}")
    
    print("\nClassification Report:")
    print(classification_report(all_labels, all_predictions, target_names=['REAL', 'FAKE']))
    
    print(f"\nConfusion Matrix:")
    print(f"                 Predicted")
    print(f"              Real   Fake")
    print(f"Actual Real   {cm[0,0]:4d}   {cm[0,1]:4d}")
    print(f"       Fake   {cm[1,0]:4d}   {cm[1,1]:4d}")
    
    # Analyze consistency patterns
    print(f"\n" + "=" * 50)
    print("CONSISTENCY ANALYSIS")
    print("=" * 50)
    
    real_indices = [i for i, label in enumerate(all_labels) if label == 0]
    fake_indices = [i for i, label in enumerate(all_labels) if label == 1]
    
    for consistency_type in ['audio_consistency', 'video_consistency', 'cross_modal_consistency']:
        real_consistency = [consistency_analysis[consistency_type][i] for i in real_indices]
        fake_consistency = [consistency_analysis[consistency_type][i] for i in fake_indices]
        
        print(f"\n{consistency_type.replace('_', ' ').title()}:")
        print(f"  Real videos - Mean: {np.mean(real_consistency):.4f}, Std: {np.std(real_consistency):.4f}")
        print(f"  Fake videos - Mean: {np.mean(fake_consistency):.4f}, Std: {np.std(fake_consistency):.4f}")
    
    # Error analysis
    print(f"\n" + "=" * 50)
    print("ERROR ANALYSIS")
    print("=" * 50)
    
    false_positives = []  # Real videos predicted as fake
    false_negatives = []  # Fake videos predicted as real
    
    for i, (pred, label, path) in enumerate(zip(all_predictions, all_labels, consistency_analysis['paths'])):
        if pred != label:
            if label == 0 and pred == 1:  # False positive
                false_positives.append((path, i))
            elif label == 1 and pred == 0:  # False negative
                false_negatives.append((path, i))
    
    print(f"False Positives (Real → Fake): {len(false_positives)}")
    if false_positives:
        print("  Sample false positives:")
        for path, idx in false_positives[:5]:
            print(f"    {path}")
    
    print(f"False Negatives (Fake → Real): {len(false_negatives)}")
    if false_negatives:
        print("  Sample false negatives:")
        for path, idx in false_negatives[:5]:
            print(f"    {path}")
    
    # Save detailed results
    if save_results:
        results = {
            'test_loss': float(test_loss),
            'test_accuracy': float(test_accuracy),
            'precision': float(precision),
            'recall': float(recall),
            'f1_score': float(f1),
            'confusion_matrix': cm.tolist(),
            'predictions': all_predictions,
            'labels': all_labels,
            'probabilities': all_probabilities,
            'consistency_analysis': consistency_analysis
        }
        
        import json
        with open('evaluation_results.json', 'w') as f:
            json.dump(results, f, indent=2)
        
        print(f"\nDetailed results saved to 'evaluation_results.json'")
    
    print("=" * 80)
    
    return {
        'accuracy': test_accuracy,
        'precision': precision,
        'recall': recall,
        'f1_score': f1,
        'confusion_matrix': cm
    }

def _get_model_outputs(model, loader, device, criterion):
    """
    Helper function to get model outputs for a given data loader.
    """
    model.eval()
    all_labels, all_probs, all_losses = [], [], []
    all_audio_features, all_video_features = [], []

    with torch.no_grad():
        for batch in tqdm(loader, desc=f"Analyzing data from {loader.dataset.mode} set"):
            audio_data = batch['audio'].to(device)
            video_data = batch['video'].to(device)
            labels = batch['label'].to(device)
            
            outputs = model(audio_data, video_data, return_features=True)
            
            # Per-sample loss using the provided criterion
            loss = criterion(outputs['logits'], labels)
            all_losses.extend(loss.cpu().numpy().tolist())
            
            probs = F.softmax(outputs['logits'], dim=1)
            
            all_labels.extend(labels.cpu().numpy().tolist())
            all_probs.extend(probs.cpu().numpy().tolist())
            
            # Aggregate features by taking the mean across the sequence length
            audio_f = torch.mean(outputs['audio_features'], dim=1)
            video_f = torch.mean(outputs['video_features'], dim=1)
            all_audio_features.append(audio_f.cpu().numpy())
            all_video_features.append(video_f.cpu().numpy())

    return {
        'labels': np.array(all_labels),
        'probs': np.array(all_probs),
        'losses': np.array(all_losses),
        'audio_features': np.concatenate(all_audio_features, axis=0),
        'video_features': np.concatenate(all_video_features, axis=0),
    }

def analyze_set_difficulty(model, val_loader, test_loader, device):
    """
    Comprehensive analysis of validation vs. test set difficulty.
    Generates visualizations for confidence, loss, and feature distributions.
    """
    print("\n" + "=" * 80)
    print("ANALYZING VALIDATION VS. TEST SET DIFFICULTY")
    print("=" * 80)

    # Use a simple criterion for per-sample loss calculation
    loss_criterion = nn.CrossEntropyLoss(reduction='none')

    # Get outputs for validation and test sets
    print("Getting model outputs for validation set...")
    val_outputs = _get_model_outputs(model, val_loader, device, loss_criterion)
    print("Getting model outputs for test set...")
    test_outputs = _get_model_outputs(model, test_loader, device, loss_criterion)
    
    # 1. Confidence Score Analysis
    val_confidence = np.max(val_outputs['probs'], axis=1)
    test_confidence = np.max(test_outputs['probs'], axis=1)

    # 2. Per-Sample Loss Analysis
    val_losses = val_outputs['losses']
    test_losses = test_outputs['losses']
    
    # 3. Class-Specific Difficulty Analysis
    val_labels = val_outputs['labels']
    test_labels = test_outputs['labels']
    val_real_conf = val_confidence[val_labels == 0]
    val_fake_conf = val_confidence[val_labels == 1]
    test_real_conf = test_confidence[test_labels == 0]
    test_fake_conf = test_confidence[test_labels == 1]
    
    # 4. Multimodal Feature Analysis
    val_audio_features = val_outputs['audio_features']
    val_video_features = val_outputs['video_features']
    test_audio_features = test_outputs['audio_features']
    test_video_features = test_outputs['video_features']

    # 5. Statistical Tests
    print("\nSTATISTICAL ANALYSIS:")
    val_mean_conf = np.mean(val_confidence)
    test_mean_conf = np.mean(test_confidence)
    t_stat, p_value = stats.ttest_ind(val_confidence, test_confidence)
    print(f"  Confidence Comparison:")
    print(f"    Validation Mean Confidence: {val_mean_conf:.4f}")
    print(f"    Test Mean Confidence: {test_mean_conf:.4f}")
    print(f"    T-test p-value: {p_value:.4f} -> {'Significant difference' if p_value < 0.05 else 'No significant difference'}")

    val_mean_loss = np.mean(val_losses)
    test_mean_loss = np.mean(test_losses)
    t_stat_loss, p_value_loss = stats.ttest_ind(val_losses, test_losses)
    print(f"  Loss Comparison:")
    print(f"    Validation Mean Loss: {val_mean_loss:.4f} (Std: {np.std(val_losses):.4f})")
    print(f"    Test Mean Loss: {test_mean_loss:.4f} (Std: {np.std(test_losses):.4f})")
    print(f"    T-test p-value: {p_value_loss:.4f} -> {'Significant difference' if p_value_loss < 0.05 else 'No significant difference'}")
    print("="*80)

    # Create visualizations
    # Main analysis plots
    fig1, axes1 = plt.subplots(2, 2, figsize=(16, 12))
    fig1.suptitle('Validation vs. Test Set Difficulty Analysis', fontsize=16, weight='bold')

    # Confidence distributions
    axes1[0, 0].hist(val_confidence, bins=30, alpha=0.7, label='Validation', density=True)
    axes1[0, 0].hist(test_confidence, bins=30, alpha=0.7, label='Test', density=True)
    axes1[0, 0].set_xlabel('Prediction Confidence')
    axes1[0, 0].set_ylabel('Density')
    axes1[0, 0].legend()
    axes1[0, 0].set_title('Confidence Score Distributions')
    axes1[0, 0].grid(True)

    # Loss distributions
    axes1[0, 1].hist(val_losses, bins=30, alpha=0.7, label='Validation', density=True)
    axes1[0, 1].hist(test_losses, bins=30, alpha=0.7, label='Test', density=True)
    axes1[0, 1].set_xlabel('Per-Sample Loss')
    axes1[0, 1].set_ylabel('Density')
    axes1[0, 1].legend()
    axes1[0, 1].set_title('Loss Distribution Comparison')
    axes1[0, 1].grid(True)

    # Class-specific confidence (REAL)
    axes1[1, 0].hist(val_real_conf, bins=20, alpha=0.7, label='Val REAL', density=True)
    axes1[1, 0].hist(test_real_conf, bins=20, alpha=0.7, label='Test REAL', density=True)
    axes1[1, 0].set_xlabel('Confidence')
    axes1[1, 0].set_ylabel('Density')
    axes1[1, 0].legend()
    axes1[1, 0].set_title('REAL Class Confidence')
    axes1[1, 0].grid(True)
    
    # Class-specific confidence (FAKE)
    axes1[1, 1].hist(val_fake_conf, bins=20, alpha=0.7, label='Val FAKE', density=True)
    axes1[1, 1].hist(test_fake_conf, bins=20, alpha=0.7, label='Test FAKE', density=True)
    axes1[1, 1].set_xlabel('Confidence')
    axes1[1, 1].set_ylabel('Density')
    axes1[1, 1].legend()
    axes1[1, 1].set_title('FAKE Class Confidence')
    axes1[1, 1].grid(True)
    
    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.savefig('set_difficulty_analysis.png', dpi=300, bbox_inches='tight')
    plt.show()
    print("\nDifficulty analysis plots saved to 'set_difficulty_analysis.png'")

    # Feature distribution plots
    fig2, axes2 = plt.subplots(1, 2, figsize=(16, 6))
    fig2.suptitle('Multimodal Feature Distribution Analysis', fontsize=16, weight='bold')

    # Audio feature distributions
    axes2[0].hist(val_audio_features.mean(axis=1), bins=30, alpha=0.7, label='Val Audio', density=True)
    axes2[0].hist(test_audio_features.mean(axis=1), bins=30, alpha=0.7, label='Test Audio', density=True)
    axes2[0].set_xlabel('Mean Audio Feature Value')
    axes2[0].set_ylabel('Density')
    axes2[0].legend()
    axes2[0].set_title('Audio Feature Distributions')
    axes2[0].grid(True)
    
    # Video feature distributions
    axes2[1].hist(val_video_features.mean(axis=1), bins=30, alpha=0.7, label='Val Video', density=True)
    axes2[1].hist(test_video_features.mean(axis=1), bins=30, alpha=0.7, label='Test Video', density=True)
    axes2[1].set_xlabel('Mean Video Feature Value')
    axes2[1].set_ylabel('Density')
    axes2[1].legend()
    axes2[1].set_title('Video Feature Distributions')
    axes2[1].grid(True)
    
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.savefig('feature_distribution_analysis.png', dpi=300, bbox_inches='tight')
    plt.show()
    print("Feature distribution plots saved to 'feature_distribution_analysis.png'")

def visualize_training_history(train_history):
    """
    Visualize training progress
    """
    print("\nGenerating training visualizations...")
    
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))
    
    # Loss curves
    axes[0, 0].plot(train_history['train_loss'], label='Train Loss', color='blue')
    axes[0, 0].plot(train_history['val_loss'], label='Validation Loss', color='red')
    axes[0, 0].set_title('Training and Validation Loss')
    axes[0, 0].set_xlabel('Epoch')
    axes[0, 0].set_ylabel('Loss')
    axes[0, 0].legend()
    axes[0, 0].grid(True)
    
    # Accuracy curves
    axes[0, 1].plot(train_history['train_acc'], label='Train Accuracy', color='blue')
    axes[0, 1].plot(train_history['val_acc'], label='Validation Accuracy', color='red')
    axes[0, 1].set_title('Training and Validation Accuracy')
    axes[0, 1].set_xlabel('Epoch')
    axes[0, 1].set_ylabel('Accuracy')
    axes[0, 1].legend()
    axes[0, 1].grid(True)
    
    # Learning rate
    axes[1, 0].plot(train_history['learning_rates'], color='green')
    axes[1, 0].set_title('Learning Rate Schedule')
    axes[1, 0].set_xlabel('Epoch')
    axes[1, 0].set_ylabel('Learning Rate')
    axes[1, 0].set_yscale('log')
    axes[1, 0].grid(True)
    
    # Best metrics summary
    best_val_acc = max(train_history['val_acc'])
    best_epoch = train_history['val_acc'].index(best_val_acc) + 1
    
    axes[1, 1].text(0.1, 0.8, f'Best Validation Accuracy: {best_val_acc:.4f}', fontsize=12, weight='bold')
    axes[1, 1].text(0.1, 0.7, f'Best Epoch: {best_epoch}', fontsize=12)
    axes[1, 1].text(0.1, 0.6, f'Final Train Accuracy: {train_history["train_acc"][-1]:.4f}', fontsize=12)
    axes[1, 1].text(0.1, 0.5, f'Final Val Accuracy: {train_history["val_acc"][-1]:.4f}', fontsize=12)
    axes[1, 1].set_title('Training Summary')
    axes[1, 1].axis('off')
    
    plt.tight_layout()
    plt.savefig('training_history.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    print("Training visualizations saved as 'training_history.png'")

# ===================================================================
# SECTION 8: MAIN EXECUTION
# ===================================================================

def main():
    """
    Main execution function
    """
    print("\n" + "=" * 80)
    print("MAIN EXECUTION - MULTIMODAL DEEPFAKE DETECTION")
    print("=" * 80)
    
    # Load datasets
    train_dataset, val_dataset, test_dataset = load_dataset()
    
    if train_dataset is None:
        print("No training data found. Please ensure videos are in the correct directories.")
        return
    
    # Create data loaders
    train_loader = DataLoader(
        train_dataset, 
        batch_size=BATCH_SIZE, 
        shuffle=True, 
        num_workers=0,  # Set to 0 for Kaggle compatibility
        pin_memory=True if device.type == 'cuda' else False
    )
    
    val_loader = DataLoader(
        val_dataset, 
        batch_size=BATCH_SIZE, 
        shuffle=False, 
        num_workers=0,
        pin_memory=True if device.type == 'cuda' else False
    )
    
    print(f"Data loaders created:")
    print(f"  Train loader: {len(train_loader)} batches")
    print(f"  Validation loader: {len(val_loader)} batches")
    
    if test_dataset is not None:
        test_loader = DataLoader(
            test_dataset, 
            batch_size=BATCH_SIZE, 
            shuffle=False, 
            num_workers=0,
            pin_memory=True if device.type == 'cuda' else False
        )
        print(f"  Test loader: {len(test_loader)} batches")
    else:
        test_loader = None
        print("  No test data available")
    
    # Initialize OPTIMIZED model
    model = MultiModalDeepfakeDetector(hidden_dim=HIDDEN_DIM, num_classes=2).to(device)
    
    # Print model info
    total_params = sum(p.numel() for p in model.parameters())
    trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
    print(f"\nModel Statistics:")
    print(f"  Total parameters: {total_params:,}")
    print(f"  Trainable parameters: {trainable_params:,}")
    print(f"  Model size: ~{total_params * 4 / (1024**2):.1f} MB")
    
    # Train the model
    train_history = train_model(model, train_loader, val_loader)
    
    # Visualize training history
    visualize_training_history(train_history)
    
    # Load best model for evaluation
    if os.path.exists('best_model.pth'):
        print("\nLoading best model for evaluation...")
        checkpoint = torch.load('best_model.pth', map_location=device, weights_only=False)
        model.load_state_dict(checkpoint['model_state_dict'])
        print(f"Loaded model from epoch {checkpoint['epoch']} with validation accuracy {checkpoint['best_val_acc']:.4f}")
    
    # Evaluate on test set if available
    if test_loader is not None:
        test_results = evaluate_model(model, test_loader)
    else:
        print("\nNo test data available for evaluation.")
        test_results = None
    
    # Analyze set difficulty
    if val_loader is not None and test_loader is not None:
        analyze_set_difficulty(model, val_loader, test_loader, device)

    # Final summary
    print("\n" + "=" * 80)
    print("EXECUTION COMPLETED SUCCESSFULLY")
    print("=" * 80)
    
    if test_results:
        print(f"Final Test Accuracy: {test_results['accuracy']:.4f}")
        print(f"Final F1-Score: {test_results['f1_score']:.4f}")
    
    print("Files generated:")
    print("  - best_model.pth (trained model)")
    print("  - training_history.png (training visualizations)")
    if test_results:
        print("  - evaluation_results.json (detailed test results)")
    print("  - set_difficulty_analysis.png (validation vs. test set comparison)")
    print("  - feature_distribution_analysis.png (feature distribution comparison)")
    
    print("\nMulti-Modal Deepfake Detection system training complete!")
    print("=" * 80)

if __name__ == "__main__":
    # Set random seeds for reproducibility
    torch.manual_seed(42)
    np.random.seed(42)
    random.seed(42)
    if torch.cuda.is_available():
        torch.cuda.manual_seed(42)
        torch.cuda.manual_seed_all(42)
    
    # Run main execution
    main()

print("\nSection 7 & 8: Evaluation and Main Execution - COMPLETED ✓")
print("=" * 80)
print("OPTIMIZED MULTIMODAL DEEPFAKE DETECTION SYSTEM READY!")
print("🚀 ENHANCED FOR SMALL DATASETS WITH ANTI-OVERFITTING MEASURES")
print("=" * 80) 
```

```python
# ===================================================================
# SECTION 9: INFERENCE-ONLY PIPELINE (CONFUSION MATRIX ON FULL DATA)
# ===================================================================

from torch.utils.data import ConcatDataset, DataLoader
from sklearn.metrics import confusion_matrix, accuracy_score, precision_recall_fscore_support
from tqdm import tqdm
import numpy as np
import torch
import os

def build_inference_dataloaders():
    """
    Reuse your load_dataset() to get train / val / test datasets,
    then build dataloaders for each + a combined FULL dataset.
    """
    print("\n" + "=" * 80)
    print("BUILDING DATASETS AND DATALOADERS FOR INFERENCE")
    print("=" * 80)

    train_dataset, val_dataset, test_dataset = load_dataset()

    loaders = {}

    if train_dataset is not None:
        loaders["train"] = DataLoader(
            train_dataset,
            batch_size=BATCH_SIZE,
            shuffle=False,
            num_workers=0,
            pin_memory=True if device.type == 'cuda' else False
        )

    if val_dataset is not None:
        loaders["val"] = DataLoader(
            val_dataset,
            batch_size=BATCH_SIZE,
            shuffle=False,
            num_workers=0,
            pin_memory=True if device.type == 'cuda' else False
        )

    if test_dataset is not None:
        loaders["test"] = DataLoader(
            test_dataset,
            batch_size=BATCH_SIZE,
            shuffle=False,
            num_workers=0,
            pin_memory=True if device.type == 'cuda' else False
        )

    # FULL DATASET = train + val (+ test if exists)
    datasets_list = [ds for ds in [train_dataset, val_dataset, test_dataset] if ds is not None]
    full_dataset = ConcatDataset(datasets_list)

    loaders["full"] = DataLoader(
        full_dataset,
        batch_size=BATCH_SIZE,
        shuffle=False,
        num_workers=0,
        pin_memory=True if device.type == 'cuda' else False
    )

    print("\nDataloaders summary for inference:")
    for k, v in loaders.items():
        print(f"  {k}: {len(v.dataset)} samples, {len(v)} batches")

    return loaders


def load_trained_model_for_inference(checkpoint_path):
    """
    Initializes the architecture and loads weights from a saved checkpoint.
    Compatible with PyTorch 2.6+ (weights_only default change).
    """
    print("\n" + "=" * 80)
    print("LOADING TRAINED MODEL FOR INFERENCE")
    print("=" * 80)

    model = MultiModalDeepfakeDetector(hidden_dim=HIDDEN_DIM, num_classes=2).to(device)

    if not os.path.exists(checkpoint_path):
        raise FileNotFoundError(f"Checkpoint not found at: {checkpoint_path}")

    # IMPORTANT: weights_only=False because this checkpoint contains more than just tensors
    checkpoint = torch.load(
        checkpoint_path,
        map_location=device,
        weights_only=False  # <- this fixes the UnpicklingError
    )

    # Handle both: full checkpoint dict OR raw state_dict
    if isinstance(checkpoint, dict) and "model_state_dict" in checkpoint:
        state_dict = checkpoint["model_state_dict"]
        print(f"Found 'model_state_dict' in checkpoint. Epoch: {checkpoint.get('epoch', 'N/A')}")
    else:
        state_dict = checkpoint
        print("Checkpoint appears to be a raw state_dict.")

    model.load_state_dict(state_dict)
    model.eval()
    print(f"Loaded model from: {checkpoint_path}")
    return model


def run_inference_on_loader(model, loader, split_name="set"):
    """
    Runs inference on a given loader and prints confusion matrix + metrics.
    Returns:
        cm, acc, precision, recall, f1
    """
    print("\n" + "=" * 80)
    print(f"RUNNING INFERENCE ON {split_name.upper()} SET")
    print("=" * 80)

    all_preds = []
    all_labels = []

    with torch.no_grad():
        for batch in tqdm(loader, desc=f"Inference [{split_name}]"):
            audio = batch["audio"].to(device)
            video = batch["video"].to(device)
            labels = batch["label"].to(device)

            outputs = model(audio, video)
            logits = outputs["logits"]
            preds = torch.argmax(logits, dim=1)

            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    all_preds = np.array(all_preds)
    all_labels = np.array(all_labels)

    cm = confusion_matrix(all_labels, all_preds)
    acc = accuracy_score(all_labels, all_preds)
    precision, recall, f1, _ = precision_recall_fscore_support(
        all_labels, all_preds, average="weighted"
    )

    # Pretty print confusion matrix
    print("\nConfusion Matrix:")
    print("                 Predicted")
    print("              Real   Fake")
    print(f"Actual Real   {cm[0,0]:4d}   {cm[0,1]:4d}")
    print(f"       Fake   {cm[1,0]:4d}   {cm[1,1]:4d}")

    print(f"\nMetrics for {split_name}:")
    print(f"  Accuracy : {acc:.4f}")
    print(f"  Precision: {precision:.4f}")
    print(f"  Recall   : {recall:.4f}")
    print(f"  F1-Score : {f1:.4f}")

    return cm, acc, precision, recall, f1


def main_inference():
    """
    MAIN ENTRY POINT FOR INFERENCE ONLY.
    - Loads datasets
    - Loads trained model from checkpoint
    - Runs inference on train / val / test / full
    - Prints confusion matrices
    """
    # 👉 CHANGE THIS to your actual Kaggle input path where best_model.pth is stored
    # Example: "/kaggle/input/hav-df-trained-model/best_model.pth"
    checkpoint_path = "/kaggle/working/best_model.pth"

    # 1. Build dataloaders (train, val, test, full)
    loaders = build_inference_dataloaders()

    # 2. Load trained model
    model = load_trained_model_for_inference(checkpoint_path)

    results = {}

    # 3. Per-split inference
    for split_name in ["train", "val", "test"]:
        if split_name in loaders:
            cm, acc, prec, rec, f1 = run_inference_on_loader(
                model, loaders[split_name], split_name=split_name
            )
            results[split_name] = {
                "cm": cm,
                "accuracy": acc,
                "precision": prec,
                "recall": rec,
                "f1": f1,
            }

    # 4. FULL DATASET confusion matrix (train + val + test)
    if "full" in loaders:
        cm_full, acc_full, p_full, r_full, f1_full = run_inference_on_loader(
            model, loaders["full"], split_name="full (train+val+test)"
        )
        results["full"] = {
            "cm": cm_full,
            "accuracy": acc_full,
            "precision": p_full,
            "recall": r_full,
            "f1": f1_full,
        }

    # (Optional) save confusion matrices
    np.save("cm_train.npy", results.get("train", {}).get("cm"))
    np.save("cm_val.npy", results.get("val", {}).get("cm"))
    if "test" in results:
        np.save("cm_test.npy", results.get("test", {}).get("cm"))
    np.save("cm_full.npy", results.get("full", {}).get("cm"))

    print("\nSaved confusion matrices as:")
    print("  cm_train.npy, cm_val.npy, cm_test.npy (if test exists), cm_full.npy")


if __name__ == "__main__":
    # Reproducibility
    torch.manual_seed(42)
    np.random.seed(42)
    random.seed(42)
    if torch.cuda.is_available():
        torch.cuda.manual_seed(42)
        torch.cuda.manual_seed_all(42)

    main_inference()
```
